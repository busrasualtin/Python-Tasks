```python
# Distribution Graphs

# Violin

import seaborn as sns
tips = sns.load_dataset("tips")
df = tips.copy()
df.head()
```

```python
sns.catplot(y="total_bill", kind="violin", data=df)
```

**Caprazlamalar**

```python
sns.catplot(x="day", y="total_bill", kind="violin", data=df)
```

```python
sns.catplot(x="day", y="total_bill", hue="sex", kind="violin", data=df)
```
