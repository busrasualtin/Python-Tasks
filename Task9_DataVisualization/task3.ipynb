```python
import seaborn as sns
planets = sns.load_dataset("planets")
df = planets.copy()
df.head()
```

```python
#are there any missing things (value) in this dataset
df.isnull().values.any()
# if the output will be true it means yes we have missing things in this dataset
```

```python
#how many missing values do we have
df.isnull().sum()
# it gives us the number of missing values for whole attributes
```

```python
# if you want to fill whole missing parts with "0"
#df["variable_name"].fillna(0, inplace=True)
df["mass"].fillna(df.mass.mean(), inplace=True)
df.isnull().sum()
```
