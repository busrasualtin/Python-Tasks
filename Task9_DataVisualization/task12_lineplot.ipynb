```python
# Distribution Graphs

# Çizgi Grafiği (Line Plot)

import seaborn as sns
fmri = sns.load_dataset("fmri")
df = fmri.copy()
df.head()
```

```python
df.shape
```

```python
df["timepoint"].describe()
```

```python
df["signal"].describe()
```

```python
#timepoint degiskenine gore grupla, neyi? signal'i ... count dedigimizde bu gruplama islemi sonrasindaki sinyalleri saymis olacak.
df.groupby("timepoint")["signal"].count()
```

Bu sonuctan anliyoruz ki her bir zaman noktasinda esit sayida sinyal toplanmis.
Eger her bir zaman noktasinda 56 tane sinyal varsa bu sinyallerin birbirinden farklı olmasını bekleriz.

```python
df.groupby("signal")["timepoint"].count()
```

```python
df.groupby("signal").count()
```

```python
#Zaman noktalari verilmis, bu zaman noktalarina gore betimsel istatistikleri toplamis olduk.
df.groupby("timepoint")["signal"].describe()
```

```python
sns.lineplot(x="timepoint", y="signal", data=df)
```

```python
sns.lineplot(x="timepoint", y="signal", hue="event", data=df)
```

```python
sns.lineplot(x="timepoint", y="signal", hue="event", style="event", data=df)
```

```python
sns.lineplot(x="timepoint",
             y="signal",
             hue="event",
             style="event",
             markers=True,
             dashes=False,
             data=df)
```

```python
sns.lineplot(x="timepoint",
             y="signal",
             hue="region",
             style="event",
             data=df)
```
