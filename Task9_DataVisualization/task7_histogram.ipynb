```python
# Distribution Graphs

# Histogram and Density

import seaborn as sns
diamonds = sns.load_dataset("diamonds")
df = diamonds.copy()
df.head()
```

```python
# I will discuss the visualization of the numerical variable named "price" with a histogram.
sns.distplot(df.price, kde=False)
# kde is related to whether the density should be placed on the graph or not.
```

```python
sns.distplot(df.price, hist=False)
```

```python
sns.kdeplot(df.price, shade=True)
```

```python
# We will create a "kdeplot" chart according to price, but we will map it with all "cut" classes using the FacetGrid function.
# To summarize again, we will create a density graph like the graph above, but we want to add dimensions on top of it.
# We want to add the classes of the "cut" categorical variable as dimensions, we want to evaluate them by crossing them. In this case, it is necessary to map sns.kdeplot with hue="cut".
(sns
 .FacetGrid(df,
               hue="cut",
               height=5,
               xlim=(0,10000))
 .map(sns.kdeplot, "price", shade=True)
 .add_legend()
)
```

```python
sns.catplot(x="cut",y="price",hue="color",kind="point",data=df)
```
