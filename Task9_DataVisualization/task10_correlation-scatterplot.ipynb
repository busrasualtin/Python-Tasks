```python
# Distribution Graphs

# Korelasyon (Correlation)
# Degiskenler arasındaki iliskiyi ifade eden istatistiksel bir terim.
# Scatterplot
# Iki degiskenin arasindaki iliskiyi ifade etmek icin kullanilan ve en cok bilinen yaklasim scatterplot yani sacilim grafigi
# Scatterplot bize sayisal degiskenler arasindaki iliskiyi gösterir

import seaborn as sns
tips = sns.load_dataset("tips")
df = tips.copy()
df.head()
```

```python
sns.scatterplot(x="total_bill", y="tip", data=df)
```

```python
sns.scatterplot(x="total_bill", y="tip", hue="time", data=df)
```

```python
#style dedigimizde time kategorik degiskeninin ozelligine gore bolunme islemi yapilmasi gerektigini anliyor
sns.scatterplot(x="total_bill", y="tip", hue="time", style="time", data=df)
```

```python
sns.scatterplot(x="total_bill", y="tip", hue="day", style="time", data=df)
```

```python
sns.scatterplot(x="total_bill", y="tip", hue="day", style="day", data=df)
```

```python
sns.scatterplot(x="total_bill", y="tip", hue="size", size="size", data=df)
```

**Dogrusal Iliskinin Gosterilmesi**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#lm=linear model
sns.lmplot(x="total_bill", y="tip", data=df)

# Dogrusal cizgi etrafindaki acik mavi bolge sapmalari gostermektedir.
# Noktaların birbirlerine cok yakin oldugu bolgede neredeyse sapma gozlenmemektedir.
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=df)
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", col="time", data=df)
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", col="time", row="sex", data=df)
```

**Scatterplot Matrisi**

```python
import seaborn as sns
iris = sns.load_dataset("iris")
df = iris.copy()
df.head()
```

```python
df.dtypes
```

```python
df.shape
```

```python
sns.pairplot(df)
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
df.groupby('species').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
figsize = (12, 1.2 * len(df['species'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(df, x='sepal_width', y='species', inner='stick', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

```python
sns.pairplot(df, hue="species")
```

```python
sns.pairplot(df, hue="species",markers=["o","s","D"])
```

```python
sns.pairplot(df, kind="reg")
```

```python
sns.pairplot(df, kind="reg", hue="species")
```
