```python
import seaborn as sns
planets = sns.load_dataset("planets")
df = planets.copy()
df.head()
```

```python
df_num = df.select_dtypes(include= ["float64", "int64"])
df_num.head()
# We only chose continuous variables.
```

```python
# We were using the describe function to access descriptive statistics.
df_num.describe().T
```

```python
df_num["distance"].describe() # We have only accessed the descriptive statistical value of the desired variable in a special way.
```

```python
print("Mean: " + str(df_num["distance"].mean()))
print("Total Num of Attr: " + str(df_num["distance"].count()))
print("Max Value: " + str(df_num["distance"].max()))
print("Min Value: " + str(df_num["distance"].min()))
print("Median: " + str(df_num["distance"].median()))
print("Standard Deviation: " + str(df_num["distance"].std()))
```
